# <font color='#7F00FF'>Formatting Output

###Concatentating Strings in print() with <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'>

Similar to Java, you can use <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> to concatenate strings in the print function. Note that the <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator does not add buffer spaces by default. So, you will need to add white space if you want separation between the strings.

<br>

**Important:** The <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator can only be used to concentate strings. So, you must be sure that every aspect being concatenated is a string. For example, the following code produces an error because it attempts to concatenate an integer.

<br>

###<font color='#0096FF'>Example

````python
# Example: This is incorrect and should produce an error

name = "Lyla Smith"
fav_num = 7

print("My name is " + name + " and my favorite number is " + fav_num) # produces an error because fav_num is not a string
````

The example above produces an error because the <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator can only concatenate strings.

<br>

To use the <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator to concatenate a value that is not a string, you must cast the variable to a string. You can fix the previous error by casting the integer to a string as shown below.

<br>

###<font color='#0096FF'>Example

````python
# Example: This is the correct way and will not produce an error

name = "Lyla Smith"
fav_num = 7

print("My name is " + name + " and my favorite number is " + str(fav_num))
````

###Concatentating Strings in print() with Commas

You can also use a comma to append strings in the print function. Unlike the <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator, the comma adds white space on either side of the value by default. The comma will also automatically cast the value to a string for you. The code below produces the same result that we got when we used the <font color='#FFFFFF'>i<font color='#000000'>+ <font color='#FFFFFF'>i<font color='#000000'> operator.

<br>

###<font color='#0096FF'>Example

````python
# Example: An alternative to using +

name = "Lyla Smith"
fav_num = 7

print("My name is",  name, "and my favorite number is", fav_num) # notice that we do not need to cast the int to a string
````

You can also use curly braces {} to pass the value of a variable to a formatted string. This makes it easier because you won't need to concatenate several different strings. Instead, you can write your variables into a single string.

<br>

###<font color='#0096FF'>Example

````python
# Example: Using {} to pass variables to a formatted string

name = "Lyla Smith"
fav_num = 7

print(f"My name is {name} and my favorite number is {fav_num}") # notice that we do not need to cast the int to a string
````

You can also use format codes to change the alignment, width, and precision of the printed values of variables of certain data types. This allows you to have more control over how variables are printed.

<br>

Format codes to specify the data type of the variable:

*   d   Decimal integer
*   s   String
*   f   Float as [-]m.dddddd
*   c   Character (from integer)

<br>

Format codes to specify the width, precision (for integers/floats), & alignment:


*   :>10d     Integer right aligned in 10-character field
*   :<10d     Integer left aligned in 10-character field
*   :^10.2d   Integer centered in 10-character field with 2 digit precision

<br>

###<font color='#0096FF'>Example

````python
stock_name = 'IBM'
shares = 100
price = 91.1
print(f'Right Aligned String: {stock_name:>10s}\nLeft Aligned Integer: {shares:<10d}\nCenter Aligned Float: {price:^10.2f}')
````

#<font color='#7F00FF'> Conditionals

In Python, conditionals allow your program to make decisions based on whether certain conditions are true or false. This allows your program to be more flexible and dynamic. For example, you might use conditionals to handle user input, to control the flow of your program, to execute loops, or to make decisions based on the results of calculations. Using conditionals effectively can make your programs more powerful and useful. There are several types of conditional statements in Python:

<br>

###<font color="#c44dff">if statements:

Used to execute code when a certain condition is true. If the condition is false, the code inside the if block is skipped.

<br>

###<font color='#0096FF'>Example

````python
# Example 1: If Statement

x = 5
if x > 0:
    print("x is positive")
````

<br>

###<font color="#c44dff"> if-else statements:

Used to execute one block of code if a certain condition is true and another block of code if the condition is false.

<br>

###<font color='#0096FF'>Examples

````python
# Example 2: If-else Statement

x = 0
if x > 0:
    print("x is positive")
else:
    print("x is negative or zero")
````

<br>

###<font color="#c44dff">if-elif-else statements:

Used to execute different blocks of code depending on multiple conditions. The first condition that is true is executed, and if none of the conditions are true, the else block is executed.

<br>

###<font color='#0096FF'>Example

````python
# Example 3: Elif Statement

x = 5
if x > 0:
    print("x is negative")
elif x == 0:
    print("x is zero")
else:
    print("x is positive")
````

<br>

###<font color="#c44dff">Ternary Operator:

The ternary operator provides a concise and readable way to write if-else statements in Python and can be useful in many situations where a condition needs to be evaluated and a value assigned based on the outcome. However, it is best used only for simple conditions and statements. If the condition and statements are too complex, it can lead to code duplication and make the code harder to maintain.


<br>

###<font color='#0096FF'>Examples

````python
#Extra: Ternary Operator --> The ternary operator is a shorthand way to write an if-else statement in a single line.

x = 5
print("x is positive" if x > 0 else "x is negative or zero")

````

<br>

###<font color="#c44dff">Nested If Statements:

You can also place conditional statements inside of each other to create nested if statements.

<br>

###<font color='#0096FF'>Example

````python
# Example 4: Nested If Statement

x = -2
if x > 0:
    if x % 2 == 0:
        print("x is positive and even")
    else:
        print("x is positive and odd")
else:
    print("x is negative or zero")
````

#<font color='#7F00FF'>Loops


###<font color="#FF00FF"> For loop:

A for loop in Python is used to iterate over a sequence of values, such as a range of numbers or characters in a string. The loop iterates over each item in the sequence and executes the code inside the loop once for each item. Here's an example of a for loop that prints the numbers from 1 to 5:

In this example, <font color="#FF00FF"> the range()</font>
function generates a sequence of numbers from 1 to 5 (inclusive), and the for loop iterates over each number in the sequence and prints it.


<br>

###<font color='#0096FF'>Example

````python
# Example: For loop that prints numbers 1 to 5

for i in range(1, 6):
    print("The value of i is:", i)
````

###<font color="#FF00FF">While loop:


A while loop in Python is used to execute a block of code repeatedly as long as a specified condition is true. The loop continues to execute as long as the condition remains true, and terminates when the condition becomes false.


<br>

###<font color='#0096FF'>Example

````python
# Example:  While loop that prints numbers from 1 to 5

i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
i = 0
sum = 0

while i < 2:
    if (i % 2) == 0:   # Line 1
      sum += i         # Line 2

    i += 1             # Line 3

print(i)               # Line 4
````

#<font color='#7F00FF'>Procedures (Function & Methods)

A <font color='red'> function </font> is a block of code that performs a specific task and returns a value. A function is defined using the def keyword, followed by the function name, any input parameters in parentheses, and a colon. The code block that makes up the body of the function is indented beneath the def statement.

<br>

###<font color='#0096FF'>Example

For example, here is a function that takes two input parameters, x and y, and returns their sum. To call this function and get its return value, you would pass two values for x and y:






````python
def add(x, y):
    return x + y

result = add(3, 5)
print(result)
````


In Python, a <font color='red'> method </font>is a type of function that is associated with an object. Methods are defined inside a class definition and are called using dot notation on an instance of the class.

<br>

###<font color='#0096FF'>Example

For example, here is a class definition that has a method named greet. To create an instance of the Person class and call its greet method, you would do the following:

````python
class Person:
    def __init__(self, name):
      self.name = name

    def set_name(self, new_name):
      self.name = new_name
      print(f"Name: {self.name}")
    def greet(self):
        print(f"Hello, my name is {self.name}!")

person = Person("Julie")
person.greet()
````
